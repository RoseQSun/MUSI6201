```python
from pyACA.computeNoveltyFunction import computeNoveltyFunction
from pyACA.ToolComputeHann import ToolComputeHann
```

```python
def computeBeatHisto(afAudioData, f_s, afWindow=None, iBlockLength=1024, iHopLength=8):

    # compute window function for FFT
    if afWindow is None:
        afWindow = ToolComputeHann(iBlockLength)

    assert(afWindow.shape[0] == iBlockLength), "parameter error: invalid window dimension"

    # novelty function
    [d, t, peaks] = computeNoveltyFunction('Flux', afAudioData, f_s, afWindow, iBlockLength, iHopLength)

    # compute autocorrelation of result
    afCorr = np.correlate(d, d, "full") / np.dot(d, d)
    afCorr = afCorr[np.arange(d.shape[0], afCorr.size)]

    Bpm = np.flip(60 / t[np.arange(1, t.shape[0])])
    T = np.flip(afCorr)

    return (T, Bpm)
```

```python
def extract_features(directory):
    data, samplerate = sf.read(directory)
    midpoint = int(len(data)/ 2)
    sect = 15 * samplerate
    data = data[midpoint - sect : midpoint + sect]
    [mfcc,t] = pyACA.computeFeature("SpectralMfccs",data,samplerate)
    [sc,t] = pyACA.computeFeature("SpectralCentroid",data,samplerate)
    [sfl,t] = pyACA.computeFeature("SpectralFlux",data,samplerate)
    [rms,t] = pyACA.computeFeature("TimeRms",data,samplerate)
    [scr,t] = pyACA.computeFeature("SpectralCrestFactor",data,samplerate)
    [zcr,t] = pyACA.computeFeature("TimeZeroCrossingRate",data,samplerate)
    features = np.stack([mfcc[0], mfcc[1],mfcc[2],mfcc[3],mfcc[4],mfcc[5],mfcc[6],mfcc[7],mfcc[8],mfcc[9],mfcc[10],mfcc[11],mfcc[12],sc, sfl, rms, scr, zcr], axis=0)
    aggFeatures = np.concatenate((np.average(features, axis=-1), np.std(features, axis=-1)),axis=0)
    return aggFeatures
```

```python
def get_feature_data(path):
    feature_data_list = []
    for directory in path:
        feature_data_list.append(extract_features(directory))
    return feature_data_list
```
